```python
# %% [markdown]
# # Module 1: Exploratory Data Analysis

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('../')

from modules.module1_data.data_loader import load_heart, load_diabetes

# %%
# Load datasets
X_heart, y_heart = load_heart()
X_diab, y_diab = load_diabetes()

# %%
# Basic statistics
print("="*50)
print("HEART DISEASE DATASET")
print("="*50)
print(f"Shape: {X_heart.shape}")
print(f"Missing values: {X_heart.isnull().sum().sum()}")
print(f"Class distribution:\n{y_heart.value_counts()}")

# %%
print("\n" + "="*50)
print("DIABETES DATASET")
print("="*50)
print(f"Shape: {X_diab.shape}")
print(f"Missing values: {X_diab.isnull().sum().sum()}")
print(f"Class distribution:\n{y_diab.value_counts()}")

# %%
# Correlation heatmap
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Heart dataset correlation
corr_heart = pd.concat([X_heart, y_heart], axis=1).corr()
sns.heatmap(corr_heart, annot=False, cmap='coolwarm', ax=axes[0])
axes[0].set_title('Heart Disease - Feature Correlations')

# Diabetes dataset correlation
corr_diab = pd.concat([X_diab, y_diab], axis=1).corr()
sns.heatmap(corr_diab, annot=False, cmap='coolwarm', ax=axes[1])
axes[1].set_title('Diabetes - Feature Correlations')

plt.tight_layout()
plt.savefig('../correlation_heatmaps.png', dpi=150, bbox_inches='tight')
plt.show()

# %%
# Distribution plots for key features
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Heart features
features_heart = ['age', 'chol', 'trestbps']
for idx, feat in enumerate(features_heart):
    axes[0, idx].hist(X_heart[feat], bins=30, alpha=0.7, color='blue')
    axes[0, idx].set_title(f'Heart: {feat}')
    axes[0, idx].set_xlabel(feat)

# Diabetes features
features_diab = ['Glucose', 'BloodPressure', 'BMI']
for idx, feat in enumerate(features_diab):
    axes[1, idx].hist(X_diab[feat], bins=30, alpha=0.7, color='green')
    axes[1, idx].set_title(f'Diabetes: {feat}')
    axes[1, idx].set_xlabel(feat)

plt.tight_layout()
plt.savefig('../feature_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```
